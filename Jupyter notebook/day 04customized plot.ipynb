```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
kashti = sns.load_dataset("titanic")
sns.boxplot(x="survived", y="age", hue="survived", showmeans=True,
            meanprops= {"marker": ".",
                       "markersize": "12",
                        "markeredgecolor": "red"},
            data=kashti)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
kashti = sns.load_dataset("titanic")
sns.boxplot(x="survived", y="age", hue="survived", showmeans=True,
            meanprops= {"marker": ".",
                       "markersize": "12",
                        "markeredgecolor": "red"},
            data=kashti)
plt.xlabel("How many survived", size=10),
plt.ylabel("Age (years)", size=10),
plt.title("Box plot of kitnay dooby and kitny bach gaye", size=14, weight="bold")
plt.show()
```
